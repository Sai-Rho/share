```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
```

```python
data=pd.read_csv('mushroom.csv')
data
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
```

```python
for col in data.columns:
    data[col] = lbl.fit_transform(data[col])
```

```python
data
```

```python
y=data['class']
x=data.iloc[:,1:23]
```

```python
import seaborn as sb
sb.pairplot(x)
```

```python
x.head()
```

```python
y.head()
```

```python
from collections import Counter
```

```python
Counter(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.cluster import KMeans
k_means=KMeans(n_clusters=2)
print(k_means.fit(x_train, y_train))
```

```python
k_means_predict = k_means.predict(x_test)
```

```python
k_means_predict
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
k_means_acc = accuracy_score(k_means_predict, y_test)
```

```python
k_means_conf = confusion_matrix(k_means_predict, y_test)
```

```python
k_means_acc
```

```python
k_means_conf
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc

```

```python
fpr, tpr, thresholds = roc_curve(y_test, k_means_predict)
```

```python
k_means_auc = auc(fpr,tpr)
```

```python

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_predict= lr.predict(x_test)
lr_predict
```

```python
lr_accuracy = accuracy_score(lr_predict, y_test)
lr_accuracy
```

```python
lr_conf = confusion_matrix(lr_predict, y_test)
lr_conf
```

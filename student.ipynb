```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

Here below we are connecting the mysql with lost host by provinding host,port,user,password and database to extract data the details are as shown below

```python
import pandas as pd
con=mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password='root',
    database='project')

data=pd.read_sql_query('select * from student',con)
data
data.head(15)
```

```python
data=data.drop(['DOB'],axis=1)
```

```python
data
```

```python
data.columns # no. of columns in dataset
```

```python
data.dtypes # data types of columns
```

```python
#converting object data type to int data type
data['Maths']=data['Maths'].astype(int)
data['Physics']=data['Physics'].astype(int)
data['Chemistry']=data['Chemistry'].astype(int)
data['English']=data['English'].astype(int)
data['Biology']=data['Biology'].astype(int)
data['Economics']=data['Economics'].astype(int)
data['History']=data['History'].astype(int)
data['Civics']=data['Civics'].astype(int)
```
